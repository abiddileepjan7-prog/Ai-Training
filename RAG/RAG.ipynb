```python

# ============================================================
# Cell 2 - Imports and Configuration
# ============================================================

import os
import uuid
import warnings

import chromadb
import numpy as np
import pandas as pd

from tqdm import tqdm

from pypdf import PdfReader
from docx import Document

import nltk
from nltk.tokenize import sent_tokenize

from sentence_transformers import SentenceTransformer

warnings.filterwarnings("ignore")

# ------------------------------------------------------------
# Download NLTK tokenizer (only runs first time)
# ------------------------------------------------------------

nltk.download("punkt", quiet=True)

# ============================================================
# Configuration
# ============================================================

# Folder containing your PDFs / DOCX / TXT files
DATA_FOLDER = "./data"

# ChromaDB storage location
CHROMA_DB_PATH = "./chroma_db"

# Collection name
COLLECTION_NAME = "oil_and_gas_rag"

# Embedding model
EMBEDDING_MODEL = "sentence-transformers/all-MiniLM-L6-v2"

# Local LLM (loaded later)
LLM_MODEL = "Qwen/Qwen2.5-1.5B-Instruct"

# Chunking
CHUNK_SIZE = 400
CHUNK_OVERLAP = 50

# Retrieval
TOP_K = 3

# ============================================================
# Initialize Embedding Model
# ============================================================

print("Loading embedding model...")

embedding_model = SentenceTransformer(EMBEDDING_MODEL)

print("Embedding model loaded.")

# ============================================================
# Initialize ChromaDB
# ============================================================

client = chromadb.PersistentClient(path=CHROMA_DB_PATH)

collection = client.get_or_create_collection(
    name=COLLECTION_NAME
)

print(f"Collection Name : {COLLECTION_NAME}")
print("ChromaDB initialized successfully.")
```

```python
# ============================================================
# Cell 3 - Read Documents
# ============================================================

documents = []

for filename in os.listdir(DATA_FOLDER):

    filepath = os.path.join(DATA_FOLDER, filename)

    # --------------------------------------------------------
    # Read PDF
    # --------------------------------------------------------
    if filename.lower().endswith(".pdf"):

        reader = PdfReader(filepath)

        text = ""

        for page in reader.pages:

            page_text = page.extract_text()

            if page_text:
                text += page_text + "\n"

        documents.append(
            {
                "filename": filename,
                "filetype": "pdf",
                "text": text
            }
        )

    # --------------------------------------------------------
    # Read DOCX
    # --------------------------------------------------------
    elif filename.lower().endswith(".docx"):

        doc = Document(filepath)

        text = "\n".join(
            paragraph.text
            for paragraph in doc.paragraphs
        )

        documents.append(
            {
                "filename": filename,
                "filetype": "docx",
                "text": text
            }
        )

    # --------------------------------------------------------
    # Read TXT
    # --------------------------------------------------------
    elif filename.lower().endswith(".txt"):

        with open(filepath, "r", encoding="utf-8") as file:

            text = file.read()

        documents.append(
            {
                "filename": filename,
                "filetype": "txt",
                "text": text
            }
        )

print(f"Loaded {len(documents)} documents.")
```

```python
# ============================================================
# Cell 4 - Clean Extracted Text
# ============================================================

import re


def clean_text(text: str) -> str:
    """
    Clean extracted document text without changing its meaning.
    """

    # Replace tabs with spaces
    text = text.replace("\t", " ")

    # Remove extra spaces
    text = re.sub(r"[ ]+", " ", text)

    # Remove excessive blank lines
    text = re.sub(r"\n{3,}", "\n\n", text)

    # Remove leading/trailing whitespace
    text = text.strip()

    return text


# Clean every document
for doc in documents:
    doc["text"] = clean_text(doc["text"])

print(f"Cleaned {len(documents)} documents.")
```

```python
# ============================================================
# Cell 5 - Recursive Pre-Chunking
# ============================================================

from nltk.tokenize import sent_tokenize


def recursive_split(text, chunk_size=400):
    """
    Recursively split text into chunks while preserving structure.
    """

    # Base case
    if len(text) <= chunk_size:
        return [text.strip()]

    # Try different separators (largest -> smallest)
    separators = [
        "\n\n",      # Paragraph
        "\n",        # Line
        ". ",        # Sentence
        " "          # Word
    ]

    for sep in separators:

        if sep in text:

            parts = text.split(sep)

            chunks = []
            current = ""

            for part in parts:

                candidate = current + part + sep

                if len(candidate) <= chunk_size:

                    current = candidate

                else:

                    if current.strip():
                        chunks.extend(
                            recursive_split(
                                current.strip(),
                                chunk_size
                            )
                        )

                    current = part + sep

            if current.strip():

                chunks.extend(
                    recursive_split(
                        current.strip(),
                        chunk_size
                    )
                )

            return chunks

    # Final fallback:
    # Hard character split
    return [
        text[i:i + chunk_size]
        for i in range(0, len(text), chunk_size)
    ]


# ------------------------------------------------------------
# Chunk every document
# ------------------------------------------------------------

all_chunks = []

for document in documents:

    chunks = recursive_split(
        document["text"],
        CHUNK_SIZE
    )

    for index, chunk in enumerate(chunks):

        all_chunks.append({

            "id": f"{document['filename']}_chunk_{index}",

            "document": document["filename"],

            "chunk_index": index,

            "text": chunk

        })

print(f"Created {len(all_chunks)} chunks.")
```

```python
# ============================================================
# Cell 6 - Sliding Window Overlap
# ============================================================

OVERLAP_SIZE = CHUNK_OVERLAP

overlapped_chunks = []

for i, chunk in enumerate(all_chunks):

    current_text = chunk["text"]

    # First chunk has no previous chunk
    if i == 0:

        final_text = current_text

    else:

        previous_text = all_chunks[i - 1]["text"]

        # Take the last OVERLAP_SIZE characters
        overlap = previous_text[-OVERLAP_SIZE:]

        final_text = overlap + "\n" + current_text

    overlapped_chunks.append({

        "id": chunk["id"],

        "document": chunk["document"],

        "chunk_index": chunk["chunk_index"],

        "text": final_text

    })

print(f"Created {len(overlapped_chunks)} overlapped chunks.")
```

```python
# ============================================================
# Cell 7 - Create Embeddings & Store in ChromaDB
# ============================================================

from tqdm import tqdm

# Optional: Clear existing data if re-running the notebook
existing = collection.count()

if existing > 0:
    print(f"Collection already contains {existing} chunks.")
    print("Deleting existing chunks...")

    ids = collection.get()["ids"]

    if ids:
        collection.delete(ids=ids)

print("Generating embeddings and storing in ChromaDB...\n")

for chunk in tqdm(overlapped_chunks):

    # Create embedding
    embedding = embedding_model.encode(
        chunk["text"],
        convert_to_numpy=True
    )

    # Store in ChromaDB
    collection.add(
        ids=[chunk["id"]],

        documents=[chunk["text"]],

        embeddings=[embedding.tolist()],

        metadatas=[

            {
                "document": chunk["document"],
                "chunk_index": chunk["chunk_index"]
            }

        ]
    )

print("\nFinished storing all chunks.")
print(f"Total Chunks in ChromaDB : {collection.count()}")
```

```python
# ============================================================
# Cell 8 - Retrieve Relevant Chunks
# ============================================================

def retrieve_chunks(query, top_k=TOP_K):
    """
    Retrieve the most relevant chunks from ChromaDB.

    Parameters
    ----------
    query : str
        User question.

    top_k : int
        Number of chunks to retrieve.

    Returns
    -------
    dict
        ChromaDB query results.
    """

    # --------------------------------------------------------
    # Convert query to embedding
    # --------------------------------------------------------

    query_embedding = embedding_model.encode(
        query,
        convert_to_numpy=True
    )

    # --------------------------------------------------------
    # Search ChromaDB
    # --------------------------------------------------------

    results = collection.query(
        query_embeddings=[query_embedding.tolist()],
        n_results=top_k
    )

    return results
```

```python
# ============================================================
# Cell 9 - Post Chunking (Context Expansion)
# ============================================================

def expand_context(results, window=1):
    """
    Expand retrieved chunks by including neighbouring chunks.

    Parameters
    ----------
    results : dict
        Output returned by retrieve_chunks()

    window : int
        Number of previous and next chunks to include.

    Returns
    -------
    str
        Expanded context for the LLM.
    """

    expanded_context = []

    # Prevent duplicate chunks
    visited = set()

    metadatas = results["metadatas"][0]

    for metadata in metadatas:

        document = metadata["document"]
        chunk_index = metadata["chunk_index"]

        # Previous -> Current -> Next
        for offset in range(-window, window + 1):

            neighbour_index = chunk_index + offset

            if neighbour_index < 0:
                continue

            neighbour_id = f"{document}_chunk_{neighbour_index}"

            # Skip duplicates
            if neighbour_id in visited:
                continue

            visited.add(neighbour_id)

            # Search in our local chunks
            for chunk in overlapped_chunks:

                if chunk["id"] == neighbour_id:

                    expanded_context.append(
                        chunk["text"]
                    )

                    break

    return "\n\n".join(expanded_context)
```

```python
# ============================================================
# Cell 9.5 - Extract Source Information
# ============================================================

def extract_sources(results):
    """
    Extract source document information from retrieved results.
    """

    sources = []
    seen = set()

    metadatas = results["metadatas"][0]
    distances = results["distances"][0]

    for metadata, distance in zip(metadatas, distances):

        key = (
            metadata["document"],
            metadata["chunk_index"]
        )

        if key not in seen:

            seen.add(key)

            sources.append({
                "document": metadata["document"],
                "chunk_index": metadata["chunk_index"],
                "distance": distance
            })

    return sources
```

```python
# ============================================================
# Cell 10 - Load Local Qwen LLM
# ============================================================

import torch

from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM
)

print("Loading Qwen Tokenizer...")

tokenizer = AutoTokenizer.from_pretrained(
    LLM_MODEL
)

print("Loading Qwen Model...")

model = AutoModelForCausalLM.from_pretrained(
    LLM_MODEL,
    torch_dtype="auto",
    device_map="auto"
)

print("Qwen loaded successfully!")

print(f"Model : {LLM_MODEL}")
```

```python
# ============================================================
# Cell 11 - Build RAG Prompt
# ============================================================

def build_prompt(question, context):
   
    prompt = f"""
You are an Oil and Gas expert assistant.

Answer the user's question ONLY using the information provided in the context.

Rules:
1. Do NOT make up information.
2. If the answer is not present in the context, reply:
   "I could not find the answer in the provided documents."
3. Keep the answer clear and concise.
4. If possible, explain in simple language.

====================================================
Context
====================================================

{context}

====================================================
Question
====================================================

{question}

====================================================
Answer
====================================================
"""

    return prompt
```

```python
# ============================================================
# Cell 12 - Generate Answer
# ============================================================

def generate_answer(question, context):

    prompt = build_prompt(question, context)

    inputs = tokenizer(
        prompt,
        return_tensors="pt"
    ).to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=300,
        temperature=0.3,
        top_p=0.9,
        do_sample=True,
        repetition_penalty=1.1,
        pad_token_id=tokenizer.eos_token_id
    )

    generated_tokens = outputs[0][inputs["input_ids"].shape[1]:]

    answer = tokenizer.decode(
        generated_tokens,
        skip_special_tokens=True
    )

    return answer
```

```python
print("=" * 70)
print("        Oil & Gas RAG Assistant")
print("=" * 70)
print("Type 'exit' to quit.\n")

while True:

    question = input("You : ").strip()

    if question.lower() == "exit":
        print("\nAssistant : Goodbye!")
        break

    if not question:
        continue

    print("\nSearching ChromaDB...")

    # --------------------------------------------------
    # Retrieve chunks
    # --------------------------------------------------

    retrieved_results = retrieve_chunks(question)

    # --------------------------------------------------
    # Expand Context
    # --------------------------------------------------

    context = expand_context(retrieved_results)

    # --------------------------------------------------
    # Extract Sources
    # --------------------------------------------------

    sources = extract_sources(retrieved_results)

    print("Generating Answer...\n")

    # --------------------------------------------------
    # Generate Answer
    # --------------------------------------------------

    answer = generate_answer(question, context)

    print("=" * 70)
    print("Assistant:\n")
    print(answer)

    print("\n" + "=" * 70)
    print("Sources Used")
    print("=" * 70)

    for i, source in enumerate(sources, start=1):

        print(
            f"{i}. {source['document']}"
        )

        print(
            f"   Chunk : {source['chunk_index']}"
        )

        print(
            f"   Distance : {source['distance']:.4f}"
        )

        print()
```
