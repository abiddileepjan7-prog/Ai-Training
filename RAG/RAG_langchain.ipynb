```python
# ============================================================
# Cell 2 - Import Required Libraries
# ============================================================

import os
from pathlib import Path

import pandas as pd
import numpy as np
from tqdm import tqdm

# -----------------------------
# LangChain Document Loaders
# -----------------------------
from langchain_community.document_loaders import (
    PyPDFLoader,
    Docx2txtLoader,
    UnstructuredHTMLLoader
)

# -----------------------------
# LangChain Document
# -----------------------------
from langchain_core.documents import Document

# -----------------------------
# Text Splitter
# -----------------------------
from langchain_text_splitters import RecursiveCharacterTextSplitter

# -----------------------------
# Embedding Model
# -----------------------------
from langchain_huggingface import HuggingFaceEmbeddings

# -----------------------------
# ChromaDB
# -----------------------------
from langchain_chroma import Chroma

# -----------------------------
# Ollama LLM
# -----------------------------
from langchain_ollama import ChatOllama

print("✅ All libraries imported successfully.")
```

```python
# ============================================================
# Cell 3 - Project Configuration
# ============================================================

# Folder containing your documents
DATA_FOLDER = "documents"

# Folder where ChromaDB will be stored
CHROMA_DB_PATH = "chroma_db"

# Embedding Model
EMBEDDING_MODEL = "sentence-transformers/all-MiniLM-L6-v2"

# Ollama Model
LLM_MODEL = "llama3.2:1b"

# Chunking Parameters
CHUNK_SIZE = 800
CHUNK_OVERLAP = 100

# Number of chunks to retrieve
TOP_K = 4

# Initialize Embedding Model
embedding_model = HuggingFaceEmbeddings(
    model_name=EMBEDDING_MODEL
)

# Initialize Ollama LLM
llm = ChatOllama(
    model=LLM_MODEL,
    temperature=0.3
)

print("✅ Configuration Loaded Successfully")
print(f"Documents Folder : {DATA_FOLDER}")
print(f"ChromaDB Folder  : {CHROMA_DB_PATH}")
print(f"Embedding Model  : {EMBEDDING_MODEL}")
print(f"LLM Model        : {LLM_MODEL}")
print(f"Chunk Size       : {CHUNK_SIZE}")
print(f"Chunk Overlap    : {CHUNK_OVERLAP}")
print(f"Top K Retrieval  : {TOP_K}")
```

```python
# ============================================================
# Cell 4 - Load Documents
# ============================================================

documents = []

# -----------------------------
# Load PDF Files
# -----------------------------
for file in Path(DATA_FOLDER).glob("*.pdf"):

    print(f"Loading PDF: {file.name}")

    loader = PyPDFLoader(str(file))
    documents.extend(loader.load())


# -----------------------------
# Load DOCX Files
# -----------------------------
for file in Path(DATA_FOLDER).glob("*.docx"):

    print(f"Loading DOCX: {file.name}")

    loader = Docx2txtLoader(str(file))
    documents.extend(loader.load())


# -----------------------------
# Load HTML Files
# -----------------------------
for file in Path(DATA_FOLDER).glob("*.html"):

    print(f"Loading HTML: {file.name}")

    loader = UnstructuredHTMLLoader(str(file))
    documents.extend(loader.load())


print("\n" + "=" * 60)
print(f"Total Documents Loaded : {len(documents)}")
print("=" * 60)
```

```python
# ============================================================
# Cell 5 - Inspect Loaded Documents
# ============================================================

print(f"Total Documents Loaded : {len(documents)}")

print("\n" + "=" * 70)
print("First Document")
print("=" * 70)

print(f"Source : {documents[0].metadata.get('source')}")

if "page" in documents[0].metadata:
    print(f"Page   : {documents[0].metadata.get('page')}")

print("\nContent Preview:\n")

print(documents[0].page_content[:1000])

print("\n" + "=" * 70)
print("Metadata")
print("=" * 70)

print(documents[0].metadata)
```

```python
# ============================================================
# Cell 6 - Chunk the Documents
# ============================================================

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=CHUNK_SIZE,
    chunk_overlap=CHUNK_OVERLAP
)

chunked_documents = text_splitter.split_documents(documents)

print(f"✅ Total Chunks Created : {len(chunked_documents)}")
```

```python
# ============================================================
# Cell 7 - Inspect Chunked Documents
# ============================================================

print(f"Total Chunks Created : {len(chunked_documents)}")

print("\n" + "=" * 70)
print("First Chunk")
print("=" * 70)

print(f"Source : {chunked_documents[0].metadata.get('source')}")

if "page" in chunked_documents[0].metadata:
    print(f"Page   : {chunked_documents[0].metadata.get('page')}")

print("\nChunk Length :", len(chunked_documents[0].page_content), "characters")

print("\nChunk Content:\n")
print(chunked_documents[0].page_content)

print("\n" + "=" * 70)
print("Metadata")
print("=" * 70)

print(chunked_documents[0].metadata)
```

```python
# ============================================================
# Cell 7 - Inspect Chunked Documents
# ============================================================

print(f"Total Chunks Created : {len(chunked_documents)}")

print("\n" + "=" * 70)
print("First Chunk")
print("=" * 70)

print(f"Source : {chunked_documents[0].metadata.get('source')}")

if "page" in chunked_documents[0].metadata:
    print(f"Page   : {chunked_documents[0].metadata.get('page')}")

print("\nChunk Length :", len(chunked_documents[0].page_content), "characters")

print("\nChunk Content:\n")
print(chunked_documents[0].page_content)

print("\n" + "=" * 70)
print("Metadata")
print("=" * 70)

print(chunked_documents[0].metadata)
```

```python
# ============================================================
# Cell 8 - Create ChromaDB Vector Store
# ============================================================

vector_db = Chroma.from_documents(
    documents=chunked_documents,
    embedding=embedding_model,
    persist_directory=CHROMA_DB_PATH
)

print("✅ ChromaDB Created Successfully")
print(f"Total Chunks Stored : {vector_db._collection.count()}")
```

```python
# ============================================================
# Cell 9 - Create Retriever
# ============================================================

retriever = vector_db.as_retriever(
    search_type="similarity",
    search_kwargs={
        "k": TOP_K
    }
)

print("✅ Retriever Created Successfully")
print(f"Top K Chunks : {TOP_K}")
```

```python
# ============================================================
# Cell 10 - Prompt Template
# ============================================================

PROMPT_TEMPLATE = """
You are a helpful AI assistant.

Answer the user's question using ONLY the provided context.

Instructions:
1. Use only the information from the context.
2. Do not make up or assume any information.
3. If the answer is not found in the context, reply:
   "I could not find the answer in the provided documents."
4. Give a clear and concise answer.

Context:
{context}

Question:
{question}

Answer:
"""

print("✅ Prompt Template Created Successfully")
```

```python
# ============================================================
# Cell 11 - RAG Function with Conversation Memory
# ============================================================

# Stores previous conversation
chat_history = []


def ask_rag(question):

    # ---------------------------------------
    # Retrieve relevant chunks
    # ---------------------------------------
    retrieved_docs = retriever.invoke(question)

    # ---------------------------------------
    # Create context
    # ---------------------------------------
    context = "\n\n".join(
        doc.page_content for doc in retrieved_docs
    )

    # ---------------------------------------
    # Include previous conversation
    # ---------------------------------------
    history = ""

    for chat in chat_history[-3:]:
        history += f"User: {chat['question']}\n"
        history += f"Assistant: {chat['answer']}\n\n"

    # ---------------------------------------
    # Create prompt
    # ---------------------------------------
    prompt = PROMPT_TEMPLATE.format(
        context=context,
        question=question
    )

    if history:
        prompt = f"Previous Conversation:\n{history}\n\n{prompt}"

    # ---------------------------------------
    # Generate answer
    # ---------------------------------------
    response = llm.invoke(prompt)

    answer = response.content

    # ---------------------------------------
    # Save conversation
    # ---------------------------------------
    chat_history.append({
        "question": question,
        "answer": answer
    })

    # ---------------------------------------
    # Print answer
    # ---------------------------------------
    print("\n" + "=" * 70)
    print("Answer")
    print("=" * 70)
    print(answer)

    # ---------------------------------------
    # Print Sources
    # ---------------------------------------
    print("\n" + "=" * 70)
    print("Sources")
    print("=" * 70)

    seen = set()

    for doc in retrieved_docs:

        source = doc.metadata.get("source", "Unknown")

        page = doc.metadata.get("page", "N/A")

        key = (source, page)

        if key not in seen:
            print(f"File : {source}")
            print(f"Page : {page}")
            print("-" * 50)

            seen.add(key)
```

```python
# ============================================================
# Cell 12 - Interactive Chat
# ============================================================

print("=" * 60)
print("        RAG Chatbot")
print("Type 'exit' to stop")
print("=" * 60)

while True:

    question = input("\nAsk a Question: ")

    if question.lower() == "exit":
        print("\nGoodbye!")
        break

    ask_rag(question)
```
